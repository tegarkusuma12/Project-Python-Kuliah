```python
# Buat dictionary kosong
data = {}

print("Masukkan pasangan key dan value.")
print("Ketik 'selesai' pada key untuk berhenti.\n")

while True:
    key = input("Key: ")
    if key.lower() == "selesai":
        break

    if key in data:
        print(f"‚ö†Ô∏è Key '{key}' sudah ada. Silakan masukkan key yang berbeda.\n")
        continue

    value = input("Value: ")

    if value in data.values():
        print(f"‚ÑπÔ∏è Value '{value}' sudah pernah dimasukkan (oleh key lain).")

    data[key] = value  # Tambahkan pasangan ke dictionary

# Cetak hasil akhir
print("\nüìã Data dalam dictionary:")
for k, v in data.items():
    print(f"{k} : {v}")
```

```python
import os

# Tentukan path file
direktori = "C:\\Users\\ASUS\\OneDrive\\Belajar-Ngoding\\Belajar-py"
nama_file = "catatan.txt"
full_path = os.path.join(direktori, nama_file)

# Fungsi: Mengecek dan menampilkan isi file
def baca_file(path):
    print("üìñ Membaca catatan sebelumnya:")
    if os.path.exists(path):
        with open(path, "r") as file:
            isi = file.read()
            print(isi if isi else "(Catatan masih kosong)")
    else:
        print("(Belum ada catatan. File akan dibuat.)")

# Fungsi: Menulis catatan baru ke file
def tulis_catatan(path):
    print("\nüìù Masukkan catatan baru:")
    catatan = input("> ")
    with open(path, "a") as file:
        file.write(catatan + "\n")
    print("‚úÖ Catatan berhasil disimpan!")

# Fungsi utama
def jalankan_program():
    print("="*30)
    print("üìÅ PROGRAM CATATAN HARIAN")
    print("="*30)

    baca_file(full_path)
    tulis_catatan(full_path)

# Panggil program
while True:
    jalankan_program()

    while True:
        sudah = input("\nApakah sudah (y/n)? : ").lower()

        if sudah == "y":
            print("Program selesai.")
            break  # keluar dari loop "apakah sudah"
        elif sudah == "n":
            break  # keluar dari loop "apakah sudah", lalu ulang jalankan_program()
        else:
            print("Ketik 'y' untuk selesai, 'n' untuk lanjut!")
    
    if sudah == "y":
        break  # keluar dari while utama juga
    
```

```python
import os

# üîπ Minta input nama file dari pengguna
while True:
    nama_file = input("Masukkan nama file (contoh: catatan.txt): ")

    # Cek apakah file sudah ada
    if os.path.exists(nama_file):
        print("‚ùå File sudah ada. Coba nama lain.")
    else:
        open(nama_file, "w").close()
        print(f"‚úÖ File '{nama_file}' siap digunakan.")
        break  # Keluar jika nama file valid dan belum ada

# üîπ Fungsi untuk membaca isi file
def baca_file():
    print("üîç Cek nama_file:", nama_file)  # Debug

    if os.path.exists(nama_file):
        with open(nama_file, "r") as file:
            isi = file.read()
            if isi.strip():
                print("\nüìñ Isi file:")
                print(isi)
            else:
                print("\nüìñ Isi file: (file kosong)")
    else:
        print("(File belum dibuat.)")

# üîπ Fungsi untuk menulis ke file
def tulis_file():
    teks = input("\nüìù Tulis catatan baru:\n> ")
    with open(nama_file, "a") as file:
        file.write(teks + "\n")
    print("‚úÖ Catatan disimpan!")

# üîπ Fungsi untuk menghapus isi file
def hapus_file():
    if os.path.exists(nama_file):
        with open(nama_file, "w") as file:
            pass  # Kosongkan isi file
        print("üóëÔ∏è Isi file berhasil dihapus!")
    else:
        print("(File tidak ditemukan.)")

# üîπ Tampilkan menu
def tampilkan_menu():
    print("\nüìã MENU:")
    print("1. Baca file")
    print("2. Tulis ke file")
    print("3. Hapus isi file")
    print("4. Keluar")

# üîπ Jalankan program
while True:
    tampilkan_menu()
    pilihan = input("Pilih menu (1/2/3/4): ")

    if pilihan == "1":
        baca_file()
    elif pilihan == "2":
        tulis_file()
    elif pilihan == "3":
        hapus_file()
    elif pilihan == "4":
        print("üëã Program selesai.")
        break
    else:
        print("‚ùå Pilihan tidak valid.")
```

```python
import os
import re

# üìÇ Direktori penyimpanan file
direktori = "C:\\Users\\ASUS\\OneDrive\\Belajar-Ngoding\\Belajar-py"
os.makedirs(direktori, exist_ok=True)

# üîπ Tampilkan semua file di direktori
def tampilkan_file():
    print("\nüìÇ Daftar file di direktori:")
    files = os.listdir(direktori)
    if not files:
        print("(Belum ada file)")
    else:
        for f in files:
            print(f"‚Ä¢ {f}")

# üîπ Buat file baru dengan validasi regex
def buat_file():
    # üìÑ Regex validasi nama file .txt
    pattern_txt = r"^[\w\s,-]+\.txt$"
    while True:
        nama = input("Masukkan nama file baru (harus .txt, atau 'batal'): ")
        if nama.lower() == "batal":
            break
        if not re.match(pattern_txt, nama):
            print("‚ùå Nama file tidak valid. Harus diakhiri .txt dan tanpa simbol aneh.")
            continue
        path = os.path.join(direktori, nama)
        if os.path.exists(path):
            print("‚ùó File sudah ada.")
            continue
        else:
            open(path, "w").close()
            print(f"‚úÖ File '{nama}' berhasil dibuat.")
            break

# üîπ Hapus file dari direktori (dengan pengulangan)
def hapus_file():
    while True:
        nama = input("Nama file yang ingin dihapus ('batal' untuk kembali): ")
        if nama.lower() == "batal":
            break
        path = os.path.join(direktori, nama)
        if os.path.exists(path):
            os.remove(path)
            print(f"üóëÔ∏è File '{nama}' berhasil dihapus.")
            break
        else:
            print("‚ùå File tidak ditemukan. Coba lagi.")

# üîπ Buka file untuk baca / tulis / hapus isi
def buka_file():
    while True:
        nama = input("Nama file yang ingin dibuka (atau ketik 'batal' untuk kembali): ")
        if nama.lower() == "batal":
            return  # kembali ke menu utama

        path = os.path.join(direktori, nama)
        if os.path.exists(path):
            break  # file ditemukan, keluar loop
        else:
            print("‚ùå File tidak ditemukan. Coba lagi.")

    while True:
        print(f"\nüìã MENU FILE: {nama}")
        print("1. Baca isi file")
        print("2. Tulis ke file")
        print("3. Hapus isi file")
        print("4. Kembali ke menu utama")
        pilih = input("Pilih (1/2/3/4): ")

        if pilih == "1":
            with open(path, "r") as f:
                isi = f.read()
                print("\nüìñ Isi file:")
                print(isi if isi else "(File kosong)")
        elif pilih == "2":
            teks = input("Tulis sesuatu: ")
            with open(path, "a") as f:
                f.write(teks + "\n")
            print("‚úÖ Tersimpan.")
        elif pilih == "3":
            open(path, "w").close()
            print("üßπ Isi file telah dihapus.")
        elif pilih == "4":
            break
        else:
            print("‚ùó Pilihan tidak valid.")

# üîπ MENU UTAMA
def menu_utama():
    while True:
        print("\nüìÅ MENU UTAMA")
        print("1. Lihat semua file")
        print("2. Buat file baru")
        print("3. Buka file")
        print("4. Hapus file")
        print("5. Keluar")
        pilih = input("Pilih (1/2/3/4/5): ")

        if pilih == "1":
            tampilkan_file()
        elif pilih == "2":
            buat_file()
        elif pilih == "3":
            buka_file()
        elif pilih == "4":
            hapus_file()
        elif pilih == "5":
            print("üëã Program selesai. Sampai jumpa!")
            break
        else:
            print("‚ùå Masukkan hanya angka 1 sampai 5.")

# üîπ Jalankan Program
menu_utama()
```
