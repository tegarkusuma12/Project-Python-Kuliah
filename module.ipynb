### Excersie 4

```python
import datetime as dt  # python standart libray

dates = dt.date(2000, 1, 1) #definisikan tanggalnya
today = dt.date.today() #ambil hari ini
days_difference = (today - dates).days #lalu menghitung selisih hari dengan hari ini 

print("Daftar tanggal:", dates)
print("Hari ini :", today)
print("Selisih hari dari hari ini:", days_difference, dates.strftime("%A")) #gunakan strftime untuk melihat waktu tertentu
```

### Excercise 5

```python
import random  # Standard Library
import numpy as np  # External Library

# Buat list angka acak
data = [random.randint(1, 100) for _ in range(5)] #generate int random 1- 100

mean_value = np.mean(data) # rata-rata
median_value = np.median(data) #median
std_value = np.std(data) #standar deviasi

# Tampilkan hasil
print("Data:", data)
print("Rata-rata:", mean_value)
print("Median:", median_value)
print("Standart Deviasi:", std_value)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([[1], [2], [3], [4], [5], [6]]) #array 2 dimensi
y = np.array([2, 4, 5, 4, 5, 7])

#model
model = LinearRegression()
model.fit(X, y)

# Prediksi nilai y berdasarkan model
y_pred = model.predict(X)

# Plot data dan hasil regresi
plt.scatter(X, y, color='blue', label="Actual Data")  # Titik data asli
plt.plot(X, y_pred, color='red', linewidth=2, label="Regression Line")  # Garis regresi

# Label dan judul grafik
plt.xlabel("X (Independent Variable)")
plt.ylabel("Y (Dependent Variable)")
plt.title("Simple Linear Regression")
plt.legend()

# Tampilkan grafik
plt.show()
```
